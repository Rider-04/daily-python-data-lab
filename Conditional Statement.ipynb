```python
# wap to check whether THE GIVEN TRIANGLE IS VALID OR INVALID,if trangle is valid then print type of triangle 
# equilateral, scalene, isoceles.
s1 = int(input("Enter the Length:- "))
s2 = int(input("Enter the Length:- "))
s3 = int(input("Enter the Length:- "))
if s1+s2>s3 and s1+s3>s2 and s2+s3>s1 :
    if s1!=s2 and s2!=s3 and s3!=s1 :
        print("This is Scalene Triangle")
    elif s1==s2 and s1==s3 and s2==s3:
        print("This is Equilateral Triangle")
    else:
        print("This is Isoceles Triangle")
else:
    print("Invalid Triangle")
```

```python
# implement a simple atm, ask user for 2 input
# first one is bank balance, second one is withdrawal amount
# update final balance ,if amount bankbalance is suffiecient 
# else insufficient bankbalance or amount
balance= int(input("Enter the Bank Balance:-"))
withdrawal= int(input("Enter the Withdrawal Amount:-"))
if balance >= withdrawal:
    print(withdrawal,"is Debited")
    balance = balance-withdrawal
    print(balance,"Remaining Balance")
else:
    print("Insifficient Bank Balance")
```

```python
# wap to check whether the given year is leap year or not
year = int(input("Enter Any Year:-"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is Leap Year")
        else:
            print(year,"is not a Leap Year")
    else:
        print(year,"is Leap Year")
else:
    print(year,"is not a Leap Year")
```
