#### Q1. Write a Python function that accepts two integer numbers. If the product of the two numbers is less than or equal to 1000,return their product; otherwise, return their sum

```python
num1 =int(input("Enter any number:-"))
num2 =int(input("Enter any number:-"))
if (num1*num2)<=1000:
    print(num1*num2)
else:
    print(num1+num2)
```

#### Q3. Display only those characters which are present at an even index number in given string.

```python
s=input("Enter any string:-")
for i in range(len(s)):
    if i%2==0:
        print(s[i],end=' ')
```

### Q4. Write a function to remove characters from a string starting from index 0 up to n and return a new string.

```python
s=input("Enter any string:-")
for i in range (len(s)):    
    print(s[i:len(s)])
```

### Q6. Write a program that calculates the factorial of a given number (e.g., 5!) using a for loop.

```python
n=int(input("Enter any number:-"))
fact=1
for i in range(1,n+1):
    fact=fact*i
print(fact)
```
