```python
# wap to check whether the given string is palindrome or not?
s=input("Enter any string:-")
rev=''
for i in s:
    rev=i+rev
if s==rev:
    print(s,'is palindrome')
else:
    print(s,'is not palindrome')
```

```python
# wpa to print number of vowels and consunants only in a given string?
s=input("Enter any string:-")
c_v=0
c_c=0
for i in range (len(s)):
    if s[i] in 'aeiouAEIOU':
        c_v=c_v+1
    elif s[i] not in 'aeiouAEIOU' and s[i] != ' ':
        c_c=c_c+1
print("There are",c_v,"vowels in the given string")
print("There are",c_c,"consunants in the given string")
print(len(s))
```

```python
# wap to find maximum and minimum character from a given string?
s=input("Enter any string:-")
print(max(s))
print(min(s))
max_s=s[:1:]
min_s=s[:1:]
for i in s:
    if i > max_s:
        max_s=i
    elif i< min_s:
        min_s=i
print(max_s)
print(min_s)
```

```python
# index() - return position of character.
# strobj.index(char/substring,[start],[stop])
# by default start ->0,stop -> end of the string. (n-1)
s='welcome to bhopal'
print(s.index('o'))
print(s.index('o',5))
print(s.index('bhopal'))
print(s.index('o',5,9))
```

```python
# find doesn't retunr error if the substring is not found 
# in the place of error it returns -1
s='welcome to bhopal'
print(s.find('o'))
print(s.find('o',5))
print(s.find('bhopal'))
print(s.find('o',5,9))
```

```python
# str.obj.replace(oldstring,newstring,[occurance])
s='welcome to bhopal'
print(s.replace('bhopal','decode'))
print(s.replace('o','a'))
print(s.replace('o','a',1))
print(s.replace('o','a',2))
print(s.replace('o','a',3))
print(s.replace('o','a',4))
```

```python
# strobj.split([separator],[maxsplit])
# by default - whitespace, maxsplit - all occurance
# it always return a list of strings.
s='welcome to bhopal'
l=s.split()
print(l)
print(s.split(' ',1))
print(s.split('o'))
```
