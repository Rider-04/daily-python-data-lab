# Broadcasting

```python
import numpy as np
```

```python
arr1=np.array([10,20,30])
arr2=np.array([10,20,30])
arr=arr1+arr2
print(arr)
```

```python
arr1=np.array([10,20,30])
arr=arr1+4
print(arr)
```

```python
arr1=np.array([10,20,30,40])
arr1=np.array([[1],[2],[3],[4]])
arr=arr1+arr2
print(arr)
```

```python
arr1=np.array([10,20,30,40])
arr2=np.array([10,20,30])
arr=arr1+arr2
print(arr)
```

```python
arr1=np.array([[10,20,30],[40,50,60]])
arr2=np.array([10,10,10])
arr=arr1+arr2
print(arr)
```

```python
arr1=np.array([[1,2],[3,4],[5,6]])
arr2=np.array([2,2,2])
arr=arr1+arr2
print(arr)
```

```python
# create one random array having 20 integers value
arr=np.random.randint(1,20,20)
print(arr)
```

```python
# create one linear spaced array having 10 values and the space between the elements
# is 5
arr=np.linspace(5,50,10)
print(arr)
```

```python
# create one array with the help of arange function having 12 values and reshape them
# into 2 and 3 dimension with all possible outcomes
arr=np.arange(1,13)
arr=np.arange(1,13).reshape(3,4)
arr=np.arange(1,13).reshape(4,3)
arr=np.arange(1,13).reshape(3,4,1)
arr=np.arange(1,13).reshape(4,3,1)
print(arr)
```
