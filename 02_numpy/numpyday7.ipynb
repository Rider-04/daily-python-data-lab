```python
import numpy as np
```

```python
arr1=np.array([10,20,30,40])
arr2=np.array([1,2,3,4])
print(arr1)
print()
print(arr2)
arr=np.concatenate((arr1,arr2))
print(arr)
```

```python
arr1=np.array([10,20,30,40])
arr2=np.array([1,2,3,4])
print(arr1)
print()
print(arr2)
arr=np.concatenate((arr1,arr2),axis=0)
print(arr)
arn=np.concatenate((arr1,arr2),axis=1)
print(arn)
```

```python
arr1=np.array([[10,20,30],[40,50,60]])
arr2=np.array([[1,2,3],[4,5,6]])
print(arr1)
print()
print(arr2)
arr=np.concatenate((arr1,arr2),axis=0)
print(arr)
arn=np.concatenate((arr1,arr2),axis=1)
print(arn)
```

```python
arr1=np.arange(1,9).reshape(2,2,2)
arr2=np.arange(1,9).reshape(2,2,2)
arr=np.concatenate((arr1,arr2),axis=0)
print(arr)
arn=np.concatenate((arr1,arr2),axis=1)
print(arn)
```

```python
arr1=np.array([10,20,30,10])
arr2=np.array([1,2,3,4])
print(arr1)
print()
print(arr2)
print()
arr=np.stack((arr1,arr2))
print(arr)
arr=np.stack((arr1,arr2),axis=0)
print(arr)
arn=np.stack((arr1,arr2),axis=1)
print(arn)
```
