```python
import numpy as np
```

```python
# datatype
arr=np.array([10,20,30,40,50],dtype='str')
print(arr)
print(arr.dtype)
```

```python
arr=np.array([10.0,20.0,30.0,40.0])
print(arr)
print(arr.dtype)
```

```python
arr1=arr.astype('int')
print(arr1)
```

```python
arr1=arr.astype(int)
print(arr1)
```

```python
# copy vs view
arr=np.array([10,20,30,40,50])
print(arr)
c=arr.copy()
c[2]=100
print(c)
print(arr)
```

```python
arr=np.array([10,20,30,40,50])
print(arr)
v=arr.view()
v[2]=100
print(v)
print(arr)
```

```python
arr=np.arange(1,10).reshape(3,3)
print(arr)
```

```python
arr=np.arange(1,10)
print(arr)
print(arr[::-1])
print(arr[2::2])
```

```python
arr=np.arange(1,10).reshape(3,3)
print(arr)
print()
print(arr[::-1])
print(arr[0][::-1])
print(arr[::-1][0])
```

```python
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
print()
print(arr[::-1])
```

```python
# arithmetic
arr=np.array([10,20,30,40])
print(arr)
arr1=arr+8
print(arr1)
arr2=np.add(arr1,8)
print(arr2)
```

```python
arr=np.array([10,20,30,40])
print(arr)
arr_new=arr-8
print(arr_new)
arr_new1=np.subtract(arr_new,2)
print(arr_new1)
```

```python
# Reciprocal
# Reciprocal works as a floor division
arr=np.array([10,20,30,40])
print(arr)
arr1=1/arr
print(arr1)
arr2=np.reciprocal(arr)
print(arr2)
```

```python
arr=np.array([10,20,30,40])
print(arr)
arr1=arr**2
print(arr1)
arr2=np.power(arr,2)
print(arr2)
```

```python
arr=np.array([10,20,30,40])
print(arr)
arr1=arr%2
print(arr1)
arr2=np.remainder(arr,2)
print(arr2)
```
