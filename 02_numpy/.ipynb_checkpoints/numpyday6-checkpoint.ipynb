```python
import numpy as np
```

```python
arr=np.array([10,2,30,0,30,402,0,103,204,30402,3012343,3204304])
print(arr)
print(np.max(arr))
print(np.argmax(arr))
```

```python
arr=np.array([10,2,30,0,30,402,0,103,204,30402,3012343,3204304])
print(arr)
print(np.min(arr))
print(np.argmin(arr))
```

```python
arr=np.array([1,4,9,16,25])
print(arr)
print(np.sqrt(arr))
```

```python
arr=np.array([1,4,9,16,25])
print(arr)
print(np.cumsum(arr))  # cumullative sum
```

```python
arr=np.array([10,20,30,40,10,302,50,30,20,0,40])
print(arr)
print(np.sort(arr))
arr1=np.sort(arr)
print(np.searchsorted(arr,9))
print(np.searchsorted(arr,[11,21,3001],side='right'))
print(np.searchsorted(arr,[11,21,3001],side='left'))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.max(arr))
print(np.argmax(arr))
print(np.max(arr,axis=0))
print(np.argmax(arr,axis=0))
print(np.max(arr,axis=1))
print(np.argmax(arr,axis=1))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.min(arr))
print(np.argmin(arr))
print(np.min(arr,axis=0))
print(np.argmin(arr,axis=0))
print(np.min(arr,axis=1))
print(np.argmin(arr,axis=1))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.sqrt(arr))
print(np.cumsum(arr,axis=0))
print(np.cumsum(arr,axis=1))
```

```python
arr=np.array([[10,1,98,56],[12,34,23,78]])
print(arr)
print(np.sort(arr))
print(np.sort(arr,axis=0))
print(np.sort(arr,axis=1))
```

```python
arr=(np.arange(1,28)*2).reshape(3,3,3)
print(arr)
print(np.sort(arr))
```

```python
# max, min, sort
arr=np.array([10,1,78,98,17])
max_value=arr[0]
min_value=arr[0]
for i in np.nditer(arr):
    if i>max_value:
        max_value=i
for i in range(len(arr)):
    if arr[i]>max_value:
        max_value=arr[i]
for i in range(len(arr)):
    if arr[i]<min_value:
        min_value=arr[i]
for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        if arr[i]>arr[j]:
            arr[i],arr[j]=arr[j],arr[i]
print(arr)
print(max_value)
print(min_value)
```
