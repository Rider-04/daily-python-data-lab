```python
# Q5. Create a dictionary:
# •	Keys → "name", "age", "city" 
# •	Values → "Parth", 21, "Bhopal" 
Keys = ["name", "age", "city"]
Values = ["Parth", 21, "Bhopal"]
d={}
for i,j in zip(Keys,Values):
    d[i]=j
print(d)
```

```python
# Q9. Print all key-value pairs using a loop.
d={'name': 'Parth', 'age': 21, 'city': 'Bhopal'}
for i in d.items():
    print(i)
```

```python
# Q10. Given:
d = {"a": 10, "b": 20, "c": 30}
# 👉 Print only keys whose values are greater than 15.
for i,j in d.items():
    if j > 15:
        print(i,end=' ')
```

```python
# Q11. Create a dictionary from two lists:
keys = ["a", "b", "c"]
values = [1, 2, 3]
d={}
for i,j in zip (keys,values):
    d[i]=j
print(d)
```

```python
# Q12. Count frequency of numbers:
lst = [2, 3, 2, 4, 3, 2, 5]
d={}
for i in lst:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print(d)
```

```python
# Q13. Count vowels in a string using a dictionary.
s=input("Enter any String:-")
d={}
for i in s:
    if i in 'aeiouAEIOU':
        if 'vowels' in d:
            d['vowels']+=1
        else:
            d['vowels']=1
print(d)
```

```python
# Q14. Remove duplicate values from a dictionary:
d = {"a": 1, "b": 2, "c": 1, "d": 3}
d1=d.copy()
v=[]
for i,j in d.items():
    if j in v:
        d1.pop(i)
    else:
        v.append(j)
print(d)
```
