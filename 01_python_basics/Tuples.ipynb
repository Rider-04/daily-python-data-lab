# Tuples

```python
t=()
print(t)
print(type(t))
```

```python
# single valued tuple
t=(5,)
print(t)
print(type(t))
```

```python
t=(10,20,30,40,'hello',10.5)
print(t)
print(type(t))
print(t[2])
t[3]=1000
print(t)
```

```python
# iteration 
# a). Direct Iterration
t=(10,20,30,40,'hello',10.5)
for i in t:
    print(i)
```

```python
t=(10,20,30,40,'hello',10.5)
for i in range(len(t)):
    print(t[i])
```

```python
# function
# len(tupleobject)
t=(10,20,30,40,'hello',10.5)
print(len(t))
count=0
for i in t:
    count=count+1
print(count)
```

```python
# max()
# min()
t=(10,20,30,40,10.5)
print(max(t))
print(min(t))
```

```python
t=(10,20,30,40,10.5)
max_value=t[0]
min_value=t[0]
for i in t:
    if i>max_value:
        max_value=i
    elif i<min_value:
        min_value=i
print(max_value)
print(min_value)
```

```python
t=(10,20,30,40,10.5)
sum=0
for i in t:
    sum=sum+i
print(sum)
```

```python
t1=(10,20,30,40)
t2=(100,200,300)
t3=t1+t2  # it create new tuple by concating the 2 or more tuples
print(t3)
```

```python
t=(10,1,19,2)
print(tuple(sorted(t,reverse=True)))
l=list(t)
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]<l[j]:
            l[i],l[j]=l[j],l[i]
print(tuple(l))
```

```python
# count()
# tupleobj.count(element)
t=(10,20,30,20,10,10,10,10.0,10.10,20.0)
print(t.count(10))
print(t.count(100))
```

```python
# index
# tupleobj.index(element,[start],[stop])
t=(10,20,30,20,10,10,10,10.0,10.10,20.0)
print(t.index(10))
print(t.index(10,4))
print(t.index(20,4,10))
```
