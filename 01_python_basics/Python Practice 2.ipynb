```python
# Q3. wap to check whether THE GIVEN TRIANGLE IS VALID OR INVALID,if trangle is valid then print type of triangle 
# equilateral, scalene, isoceles.
s1=int(input("Enter any number:-"))
s2=int(input("Enter any number:-"))
s3=int(input("Enter any number:-"))
if s1+s2>s3 and s1+s3>s2 and s2+s3>s1:
    if s1==s2==s3:
        print("Equilateral Triangle")
    elif s1!=s2!=s3:
        print("Scalene Triangle")
    else:
        print("Isosceles Triangle")
else:
    print("Not A Triangle")
```

```python
# Q5.implement a simple atm, ask user for 2 input
# first one is bank balance, second one is withdrawal amount
# update final balance ,if amount bankbalance is suffiecient 
# else insufficient bankbalance or amount
bb=int(input("Enter the Bank Balance:-"))
wa=int(input("Enter the Withdrawal Amount:-"))
if wa<=bb:
    bb=bb-wa
    print(wa,'amount is withdrawaled from account')
    print(bb,'remaining balance')
else:
    print("insufficient balance")
```

```python
# Q11.wap to print sum of first n even number and odd number?
n=int(input('Enter any number:-'))
sum_e=0
sum_o=0
for i in range(1,(2*n)+1):
    if i%2==0:
        sum_e=sum_e+i
    else:
        sum_o=sum_o+i
print("Sum of first",n,"even number is",sum_e)
print("Sum of first",n,"odd number is",sum_o)
```

```python
# Q12.wap to print factorial of any number 
n=int(input("Enter any number:-"))
fact=1
if n==0:
    print("0 is the factorial of",n)
else:
    for i in range(1,n+1):
        fact=fact*i
    print(fact,'is factorial of',n)
```

```python
# Q19.wap to check whether the given number is an armstrong or not?
# armstrong is the number in which the number and the sum of the individual number with the exponential of there
# length is equal 
n= int(input("Enter any number:-"))
old=n
exp=len(str(n))
arm=0
while n!=0:
    dig=n%10
    arm=arm+(dig**exp)
    n=n//10
if old==arm:
    print(old, "is armstrong number")
else:
    print(old, "is not armstrong number")
```

```python
# Q20. wap to print armstrong number between 1 to 2000?
start=int(input("Enter any number:-"))
end=int(input("Enter any number:-"))
for i in range (start,end+1):
    old=i
    arm=0
    exp=len(str(i))
    while i!=0:
        dig=i%10
        arm=arm+(dig**exp)
        i=i//10
    if arm==old:
        print(arm,end=' ')
```
