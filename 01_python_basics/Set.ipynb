```python
# pop
# setobj.pop()
s={10,20,10,20,'hello',10.5}
print(s.pop())
print(s)
```

```python
# remove
# setobj.remove(element)
s={10,20,10,20,'hello',10.5}
s.remove(10.5)
print(s)
s.remove(1000)
print(s)
```

```python
# discard
# setobj.discard(element)
# difference between remove and discard is just that , remove gives error when element not found 
# but discard doesnt gives the error when element not found
s={10,20,10,20,'hello',10.5}
s.discard(1000)
print(s)
```

```python
# update
# setobj.remove(iterables)
s={10,20,10,20,'hello',10.5}
s.update([10,20,30,40])
print(s)
```

```python
# wap that print the given number is prime or not.
n=int(input("Enter any number:-"))
if n!=1:
    for i in range(2,n):
        if n%i==0:
            print(n,'is not a prime number')
            break
    else:
        print(n,'is prime number')  
else:
    print(n,'is not a prime number')
```

```python
# wap that print febonacci series upto nth term.
n=int(input("Enter any number:-"))
f=0
s=1
if n==1:
    print(f)
elif n==2:
    print(f,s)
else:
    print(f,s,end=' ')
    for i in range(2,n):
        c=f+s
        print(c,end=' ')
        f=s
        s=c
```

```python
# wap that print armstrong number between start and end .
start=int(input("Enter the start:-"))
end=int(input("Enter the end:-"))
for i in range(start,end+1):
    if i>0:
        a=len(str(i))
        old=i
        sm=0
        while i!=0:
            dig=i%10
            sm=sm+(dig**a)
            i=i//10
        if old==sm:
            print(old,end=' ')
```
