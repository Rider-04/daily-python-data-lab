```python
s=set()
print(s)
print(type(s))
l=[10,20,30,10,10,'hello']
s=set(l)
print(s)
```

```python
# setobj.add(value)
l=[10,20,30,10,10,'hello']
s=set(l)
print(s)
s.add('h')   # mutable
print(s)
```

```python
# iteration
s={10,20,10,20,'hello',10.5}
for i in s:
    print(i)
```

```python
s={10,20,10,20,'hello',10.5}
print(len(s))
count=0
for i in s:
    count=count+1
print(count)
```

```python
s={10,20,10,20,10.5}
print(max(s))
print(min(s))
```

```python
# pop
# setobj.pop()
s={10,20,10,20,'hello',10.5}
print(s.pop())
print(s)
```

```python
# remove
# setobj.remove(element)
s={10,20,10,20,'hello',10.5}
s.remove(10.5)
print(s)
s.remove(1000)
print(s)
```

```python
# discard
# setobj.discard(element)
# difference between remove and discard is just that , remove gives error when element not found 
# but discard doesnt gives the error when element not found
s={10,20,10,20,'hello',10.5}
s.discard(1000)
print(s)
```

```python
# update
# setobj.remove(iterables)
s={10,20,10,20,'hello',10.5}
s.update([10,20,30,40])
print(s)
```
