```python
l=[3,2,5,7,4,6]
d={}
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]>l[j]:
            l[i],l[j]=l[j],l[i]
for i in l:
    if i%2==0:
        d[i]=i**3
print(d)
```

```python
n=int(input('Enter any number:-'))
d={}
for i in range(n):
    c=65
    c=c+i
    if c%2!=0:
        d[chr(c)]=(i+1)**3
print(d)
```

```python
l=['red','brown','red','green','red','brown','yellow','Yellow']
l=['red','brown','red','green','red','brown','yellow','Yellow']
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print(d)
```

```python
l1=[5,1,6,2,4,3]
l2=[2,6,3,5,1,4]
d={}
# output:- {6:216,2:32,4:4}
for i in range(len(l1)):
    if l1[i]%2==0:
        d[l1[i]]=l1[i]**l2[i]
print(d)
```

```python
l1=[5,1,6,2,4,3]
l2=[2,6,3,5,1,4]
d={}
# output:- {6:216,2:32,4:4}
for i,j in zip(l1,l2):
    if i%2==0:
        d[i]=i**j
print(d)
```

```python
# methods
# dictobj.get(key)
d={"name":["abc","def","pqr"],"subject":["DA","DS","DE"],"fees":[30000,40000,40000]}
print(d)
print(d.get('name'))
for i in d.get('name'):
    print(i)
```

```python
# dictobj.keys()
d={"name":["abc","def","pqr"],"subject":["DA","DS","DE"],"fees":[30000,40000,40000]}
print(d.keys())
for i in d.keys():
    print(i)
```

```python
# dictobj.values()
d={"name":["abc","def","pqr"],"subject":["DA","DS","DE"],"fees":[30000,40000,40000]}
print(d.values())
for i in d.values():
    print(i)
```

```python
# dictobj.items()
d={"name":["abc","def","pqr"],"subject":["DA","DS","DE"],"fees":[30000,40000,40000]}
print(d.items())
for i,j in d.items():
    print(i,j)
```

```python
# mean,median,mode
l=[2,2,1,4,3,5,6,4,5,6,7,7,7,7,7,7,7,7,3,2,6,9,8,10,11,1,5,7]
l1=sorted(l)
d={}
print(l1)
n=len(l1)
sum=0
count=0
for i in l:
    sum=sum+i
    count=count+1
mean=sum/len(l)
mean1=sum/count
print(mean,mean1)
median=(l1[(n//2)-1]+l1[(n//2)])/2
print(median)
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
val=d.values()
m=max(val)
for i,j in d.items():
    if j==m:
        print(i)
```

```python
words=['data','science','AI','python','data','code','python']
d={}
for i in words:
    if len(i)>3:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
print(d)
```

```python
marks={
    "Ankit":90,
    "Ravi":85,
    "Neha":90,
    "Aman":80
}
d={}
for i,j in marks.items():
    if j not in d:
        d[j]=[]
    d[j].append(i)   
print(d)
```

```python
marks={
    "Ankit":90,
    "Ravi":85,
    "Neha":90,
    "Aman":80
}
d={}
for i,j in marks.items():
    if j in d:
        d[j].append(i)
    else:
        d[j]=[i]
print(d)
```

```python
nums=[10,15,21,22,30,33,40]
d={}
for i in nums:
    if i%3==0:
        if i%2==0:
            d[i]='even'
        else:
            d[i]='odd'
print(d)
```

```python
# dictonary comprehension
# consice way to create a dictonary
# {k:v looping statement conditional statement}
l=[1,2,3,4,5]
d={}
for i in l:
    d[i]=i**3
print(d)
print({i:i**3 for i in l})
```

```python
l=[10,1,5,6,7,9]
# {1,25,49,81}
d={}
for i in l:
    if i%2!=0:
        d[i]=i**2
print(d)
print({i:i**2 for i in l if i%2!=0})
```

```python
words=['data','science','AI','python','data','code','python']
# {'data': 2, 'science': 1, 'python': 2, 'code': 1}
d={}
for i in words:
    if len(i)>3:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
print(d)
print({i:words.count(i) for i in words if len(i)>3})
```

```python
nums=[10,15,21,22,30,33,40]
d={}
for i in nums:
    if i%3==0:
        if i%2==0:
            d[i]='even'
        else:
            d[i]='odd'
print(d)
print({i:"even" if i%2==0 else"odd" for i in nums if i%3==0})
```
