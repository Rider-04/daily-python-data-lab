# While Loop

```python
# wap to print first n natural number
n=int(input("Enter any Number:-"))
for i in range(1,n+1):
    print(i,end=" ")
```

```python
# wap to print first n even number and odd number
n=int(input("Enter any Number:-"))
for i in range(1,2*n+1):
    if i%2==0:
        print(i,end=" ")
    print()
    if i%2!=0:
        print(i,end=" ")
```

```python
# wap to print factorial of n 
n=int(input("Enter any number:-"))
fact=1
for i in range(1,n+1):
    fact=fact*i
print(fact)
```

```python
# wap to print sum of first n even number and odd number
n=int(input("Enter any number:-"))
sum_e=0
sum_o=0
for i in range(2*n+1):
    if i%2==0:
        sum_e=sum_e+i
    else:
        sum_o=sum_o+i
print(sum_e,"is sum of first n even number")
print(sum_o,"is sum of first n odd number")
```

```python
# wap to print first n natural number
n = int(input("Enter any number:-"))
i=1
while i<n+1:
    print(i,end=" ")
    i=i+1
```

```python
# wap to print first n even number and odd number
n = int(input("Enter any number:-"))
i=1
while i<2*n+1:
    if i%2==0:
        print(i,"even number")
    else:
        print(i,"Odd number")
    i=i+1
```

```python
# wap to print factorial of n 
n=int(input("Enter any number:-"))
fact=1
i=1
while i<n+1:
    fact=fact*i
    i=i+1
print(fact)
```

```python
# wap to print sum of first n even number and odd number
n=int(input("Enter any number:-"))
i=1
sum_e=0
sum_o=0
while i<2*n+1:
    if i%2==0:
        sum_e=sum_e+i
    else:
        sum_o=sum_o+i
    i=i+1
print(sum_e,"is sum of first n even number")
print(sum_o,"is sum of first n odd number")
```

```python
# wap to print fibonacci series up to n'th term?
n=int(input("Enter any number:-"))
a=0
b=1
i=2
print(a,b,end=' ')
while i<n:
    c=a+b
    a=b
    b=c
    print(c,end=" ")
    i=i+1
```

```python
# wap to print n'th term of fibonacci?
n=int(input("Enter any number:-"))
a=0
b=1
i=2
while i<n:
    c=a+b
    a=b
    b=c
    i=i+1
print(c)
```

```python
# wap to check wehther the given number is prime or not?
n=int(input("Enter any number:-"))
i=1
if n==0 or n==1:
    print(n,"not prime number")
elif n>1:
    while i<n:
        if n%i==0:
            print(n,"prime number")
            break
        else:
            print(n,"not prime number")
        i=i+1
else:
    print(n,"not prime number")
```

```python
# wap to print reverse of any given number
n=int(input("Enter any number:-"))
rev=0
while n!=0:             # n>0 is also acceptable
    dig=n%10            # last digit
    rev=(rev*10)+dig    # output exp
    n=n//10             # loop
print(rev)
```

```python
# wap to check whether the given number is an armstrong or not?
n=int(input("Enter any number:-"))
a=len(str(n))
old=n
sm=0
while n!=0:
    dig=n%10
    sm=sm+(dig**a)
    n=n//10
if sm==old:
    print(old,"is Armstrong Number")
else:
    print(old,"is not a Armstrong Number")
```

```python
# wap to print armstrong number between 1 to 2000?
start=int(input("Enter any number:-"))
end=int(input("Enter any number:-"))
for i in range (start,end+1):
    a=len(str(i))
    old=i
    sm=0
    while i!=0:
        dig=i%10
        sm=sm+(dig**a)
        i=i//10
    if sm==old:
        print(old,end=" ")
```

```python
n= int(input("Enter any number:-"))

i=0
j=0
while i<n:
    while j<n:
        print('*',end=' ')
        j=j+1
    i=i+1
    print()
print()
```
