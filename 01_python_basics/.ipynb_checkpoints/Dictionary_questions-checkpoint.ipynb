```python
# Q1. Nested Dictionary Update
# Given the following dictionary:
students = {"S1": {"Python": 70, "SQL": 60, "PowerBI": 40},
            "S2": {"Python": 85, "SQL": 70, "PowerBI": 30},
            "S3": {"Python": 60, "SQL": 55, "PowerBI": 50}
}
# Increase the Python marks by 5 only for students whose total marks are less than 180.
# Modify the dictionary in-place.
for i in students:
    if (sum(students[i].values()))<180 :
        students[i]['Python']+=5
print(students)
```

```python
# Q2. Sales Aggregation
# Given sales data:
sales = [{"product": "Laptop", "amount": 50000},
         {"product": "Mobile", "amount": 20000},
         {"product": "Laptop", "amount": 55000},
         {"product": "Tablet", "amount": 15000},
         {"product": "Mobile", "amount": 25000}
]
d={}
# Write a program to calculate total sales per product using a dictionary.
for i in sales:
    p=i['product']
    a=i['amount']
    if p not in d:
        d[p]=0
    d[p]+=a
print(d)
```

```python
# Q3. Given a list of numbers:
nums = [10, 20, 30, 40, 50]
l = [10, 20, 30, 40, 50]
# Create a dictionary where:
# - Keys are indices
# - Values are the average of the last three elements (including current)
# - If fewer than three elements exist, take average of available elements
d={}
for i in range(len(l)):
    if i==0:
        avg=l[i]/1
    elif i==1:
        avg=(l[i]+l[i-1])/2
    else:
        avg=(l[i]+l[i-1]+l[i-2])/3
    d[i]=avg
print(d)
```

```python
# Dictionary question
# Q1.Given two dictionaries:
d1 = {"A": 10,
    "B": {"x": 5, "y": 3},
    "C": 7}
d2 = {"A": 5,
    "B": {"x": 2, "z": 9},
    "D": 4}
# Merge these dictionaries recursively using the following rules:
# - If both values are dictionaries, merge them
# - If both values are integers, add them
# - Otherwise, overwrite with value from d2
for i,j in d1.items():
    if i in d2:
        if type(d1[i])==int and type(d2[i])==int:
            d1[i]=j+d2[i]
        elif type(d1[i])==dict and type(d2[i])==dict:
            d1[i].update(d2[i])
    for a,b in d2.items():
        if a not in d1:
            d1[a]=d2[a]
print(d1)
```

```python
def merge_dicts(d1, d2):
    d3 = {}
    for i in d1:
        if i in d2:
            if (type(d1[i])==dict) and (type(d2[i])== dict):
                d3[i] = merge_dicts(d1[i], d2[i])
            elif (type(d1[i])==int) and (type(d2[i])== int):
                d3[i] = d1[i] + d2[i]
            else:
                d3[i] = d2[i]
        else:
            d3[i] = d1[i]
    for i in d2:
        if i not in d1:
            d3[i] = d2[i]
    return d3
d1 = {"A": 10,
      "B": {"x": 5, "y": 3},
      "C": 7}
d2 = {"A": 5,
      "B": {"x": 2, "z": 9},
      "D": 4}
merged = merge_dicts(d1, d2)
print(merged)
```

```python
# Q2. 
# Given a dictionary of subject scores:
scores = {
    "Math": 78,
    "Python": 92,
    "SQL": 85,
    "PowerBI": 70
}
# Normalize the values between 0 and 1 using min-max normalization:
# (x - min) / (max - min)
# Store the result in a new dictionary and round values to 2 decimal places.

d={}
max = max(scores.values())
min= min(scores.values())
for i,j in scores.items():
    d[i]= (j - min)/(max - min)
print(d)
    
```

```python
# Q2. 
# Given a dictionary of subject scores:
scores = {
    "Math": 78,
    "Python": 92,
    "SQL": 85,
    "PowerBI": 70
}
# Normalize the values between 0 and 1 using min-max normalization:
# (x - min) / (max - min)
# Store the result in a new dictionary and round values to 2 decimal places.

d={}
max_val = max(scores.values())
min_val = min(scores.values())
for i,j in scores.items():
    d[i]= round((j - min_val) / (max_val - min_val), 2)
print(d)
    
```
