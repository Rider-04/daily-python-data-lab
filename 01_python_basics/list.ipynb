## sorted()

```python
# sorted function -sorted(listobj,[reverse=false])
l=[10,2,4,7,1]
l1=sorted(l)
print(l1) # asc order
l2=sorted(l,reverse=True)
print(l2) # decs order
```

```python
l=[10,5,6,7,9,8,9,7,2,3,3,3,4,5,6,3,9,3]
n=len(l)
sorted_list=sorted(l)
sum=0
count=0
for i in l:
    sum=sum+i
    count=count+1
mean=sum/count
print(mean)
if n%2!=0:
    median=sorted_list[n//2]
else:
    median=(sorted_list[n//2]+sorted_list[n//2-1])/2
print(median)
```

### pop
### del
### remove
### clear

```python
# pop
# listobj.pop([position])
# by default - last element (-1)
# pop method returns element which is being removed
l=[10,20,30,'hello',100.5]
print(l.pop())
print(l)
l.pop(2)
print(l)
```

```python
# del
# del listobj[position]
l=[10,20,30,40,'hello',100.5]
del l[3]
print(l)
```

```python
# remove
#listobj.remove(element)
l=[10,20,30,'hello',10,100.5]
l.remove(10)
print(l)
```

```python
# clear
# listobj
l=[10,20,30,'hello',100.5]
l.clear()
print(l)
```

```python
# index
# listobj.index(element,[start],[stop])
# by default start-0 and stop - last element
l=[10,20,30,'hello',10,20,100,10,20,'decode']
print(l.index(10))
print(l.index(10,1))
print(l.index(10,1,4))
```

```python
# count
# listobj.count(element)
l=[10,20,30,'hello',10,20,100,10,20,'decode']
print(l.count(10))
print(l.count('h'))
print(l.count(10,1))
```

```python
# zip function
l1 = [10,20,30,40,500,600,700]
l2 = [100,200,300,400,1000,2000,3000,4000,5,90]
for i,j in zip(l1,l2):
    print(i,j)
```

```python
# enumerate()
l=[10,20,30,40,50]
for i in enumerate(l):  # in i the list of position and value is printed
    print(i)
for i,j in enumerate(l):  # i = position and j = value
    print(i,j)
```

# List Comprehension

```python
# syntax:-
# [expression looping statement conditional statement]
l=[1,2,3,4,5]
l1=[]
for i in l:
    l1.append(i**3)
print(l1)
print([i**3 for i in l])
```

```python
l = [18,2,17,6,12,19]
l1=[]
for i in l:
    if i%2==0:
        l1.append(i)
print(l1)
print([i for i in l if i%2==0])
```

```python
l1=[6,4,1,3,2,5]
l2=[1,3,6,4,5,2]
l3=[]
for i,j in zip(l1,l2):
    l3.append(i**j)
print(l3)
print([i**j for i,j in zip(l1,l2)])
```

```python
l=[2,7,5,10,3,9,8,12,19]
l1=[]
for i in l:
    if i%2==0:
        l1.append(i**2)
print(l1)
print([i**2 for i in l if i%2==0])   
```

```python
l=['madam','bhopal','Naman','decode','nitin']
l1=[]
for i in l:
    if i!=i[::-1]:
            l1.append(i)
print(l1)
print([i for i in l if i!=i[::-1]])
```
