# DataTypes

### Numeric

```python
# int
# type() - returns datatype of a given object
a=10
print(a)
print(type(a))
```

```python
# float
a=10
b=5
c=a/b
print(c)
print(type(c))
```

### String

```python
# string- ordered sequence of characters
# it is denoted by '', ""
# multi-line string is denoted by ''' '''
# indexed , immutable
s=''
print(s)
print(type(s))
```

```python
s="hello"
print(s)
print(type(s))
print(s[0])  # indexed
s[0] = 'z'   # immutable
print(s)
```

### Sequence

### List

```python
# list-collection of elements
# it is denoted by []
# ordered,indexed,mutable and heterogenous in nature
l=[]  # empty list
print(l)
print(type(l))
```

```python
l=[10,20,30,"hello",10.5] 
print(l)
print(l[1]) # indexed
print(type(l))
l[3]='decode' # mutable
print(l)
```

### Tuple

```python
# collection of elements
# denoted by () with csv
# orddered,indexed,immutable and heterogenous in  nature.
t=()  # empty tuple
print(t)
print(type(t))
```

```python
# single valued tuple
t=(10,) #single valued tuple
print(t)
print(type(t))
```

```python
t=(10,20,30,"hello",10.5)
print(t)
print(type(t))
print(t[0])  # indexed
t[0] = 1000  # immutable
print(t)
```

### Dictionary

```python
# collection of element
# denoted by {k:v} and data stored in key:value pair
# keys must be unique and values can be duplicate
# ordered,unindexed and mutable.
d={}   # empty dictinoray
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"da","fees":30000}
print(d)
print(d["subject"])  # unindexed
d["Name"] = "abc"    # mutable and values can be duplicate
d["name"]="decode"   # keys must be unique
print(d)
```

### Set

```python
# collection of unique elements.
# denoted by {} with csv
# unrdered,unindexed,mutable and heterogenous
s=set()  # empty set
print(s)
print(type(s))
s.add(1000)
print(s)
```

```python
s={10,20,30,"hello|",10.5,10,20,10,20,10}
print(s)
print(type(s))
s.add(1000)
print(s)
```

### Frozenset

```python
# collection of unique elements
# denoted by constructor i.e frozenset()
# unordered,unindexed,immutable and heterogenous
f=frozenset()
print(f)
print(type(f))
```

```python
f=frozenset({10,20,10,20,"hello",10.5,20,10})
print(f)
print(type(f))
```
