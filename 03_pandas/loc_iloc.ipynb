```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':['Parth','Priyanshu','Rishabh','Arman','Suraj',"Aman"],
                'Age':[21,22,23,24,23,21],
                'City':['Bhopal','Raisen','Indore','Gwalior','Jabalpur','Raisen']},index=['a','b','c','d','e','f'])
df
```

```python
# loc - label based indexing
# loc[rows,columns]
df.loc[['a','b','d']]
```

```python
df.loc[['a','b','d'],['Name']]
```

```python
df.loc[::,['Name',"City"]]
```

```python
df.loc[df['City']=='Raisen']
```

```python
df.loc[:'d':,:'Age':]
```

```python
df.iloc[0::2,[0,-1]]
```

```python
df.iloc[:4:,::]
```

```python
df.loc[:"e":,::]
```
