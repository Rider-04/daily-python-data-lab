```python
Q2=np.quantile(df_prev["AMT_ANNUITY"],0.50)
Q2
```

```python
Q3=np.quantile(df_prev["AMT_ANNUITY"],0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_fence=Q1-(1.5*IQR)
lower_fence
```

```python
upper_fence=Q3+(1.5*IQR)
upper_fence
```

```python
l=[]
for i in df_prev["AMT_ANNUITY"]:
    if i<lower_fence or i>upper_fence:
        l.append(i)
print(len(l))
```
