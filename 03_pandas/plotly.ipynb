```python
#started Learning Plotly
# ============================================================
# PLOTLY COMPLETE PRACTICE NOTEBOOK / SCRIPT
# Beginner -> Intermediate -> Advanced
# ============================================================

# If Plotly is not installed, run:
# pip install plotly pandas numpy

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. SAMPLE DATASETS
# ------------------------------------------------------------

# Built-in Plotly datasets
df_gapminder = px.data.gapminder()
df_tips = px.data.tips()
df_iris = px.data.iris()
df_penguins = px.data.penguins()
df_stocks = px.data.stocks()
df_medals = px.data.medals_long()

# Custom dataset
np.random.seed(42)
sales_df = pd.DataFrame({
    "Month": pd.date_range("2025-01-01", periods=12, freq="M"),
    "Sales": np.random.randint(20000, 80000, 12),
    "Profit": np.random.randint(5000, 25000, 12),
    "Region": np.random.choice(["North", "South", "East", "West"], 12),
    "Category": np.random.choice(["Electronics", "Clothing", "Furniture"], 12),
    "Orders": np.random.randint(100, 500, 12)
})

print("Data Loaded Successfully!")
print("sales_df preview:")
print(sales_df.head())

# ------------------------------------------------------------
# 2. PLOTLY EXPRESS BASICS
# ------------------------------------------------------------
# Plotly Express = quick and easy plotting
# Syntax style:
# px.chart_type(data_frame, x=..., y=..., color=..., title=...)

# ------------------------------------------------------------
# 3. LINE CHART
# ------------------------------------------------------------
fig = px.line(
    sales_df,
    x="Month",
    y="Sales",
    title="Monthly Sales Trend"
)
fig.show()

# Multiple lines
fig = px.line(
    sales_df,
    x="Month",
    y=["Sales", "Profit"],
    title="Sales and Profit Trend"
)
fig.show()

# ------------------------------------------------------------
# 4. BAR CHART
# ------------------------------------------------------------
fig = px.bar(
    sales_df,
    x="Region",
    y="Sales",
    color="Category",
    title="Sales by Region and Category",
    barmode="group"
)
fig.show()

# Stacked bar
fig = px.bar(
    sales_df,
    x="Region",
    y="Sales",
    color="Category",
    title="Stacked Bar Chart",
    barmode="stack"
)
fig.show()

# ------------------------------------------------------------
# 5. SCATTER PLOT
# ------------------------------------------------------------
fig = px.scatter(
    sales_df,
    x="Sales",
    y="Profit",
    color="Category",
    size="Orders",
    hover_data=["Region"],
    title="Sales vs Profit"
)
fig.show()

# ------------------------------------------------------------
# 6. HISTOGRAM
# ------------------------------------------------------------
fig = px.histogram(
    df_tips,
    x="total_bill",
    nbins=20,
    color="sex",
    title="Histogram of Total Bill"
)
fig.show()

# ------------------------------------------------------------
# 7. BOX PLOT
# ------------------------------------------------------------
fig = px.box(
    df_tips,
    x="day",
    y="total_bill",
    color="sex",
    title="Box Plot of Total Bill by Day"
)
fig.show()

# ------------------------------------------------------------
# 8. VIOLIN PLOT
# ------------------------------------------------------------
fig = px.violin(
    df_tips,
    x="day",
    y="total_bill",
    color="sex",
    box=True,
    points="all",
    title="Violin Plot"
)
fig.show()

# ------------------------------------------------------------
# 9. PIE CHART
# ------------------------------------------------------------
region_sales = sales_df.groupby("Region", as_index=False)["Sales"].sum()

fig = px.pie(
    region_sales,
    names="Region",
    values="Sales",
    title="Region-wise Sales Distribution"
)
fig.show()

# ------------------------------------------------------------
# 10. DONUT CHART
# ------------------------------------------------------------
fig = px.pie(
    region_sales,
    names="Region",
    values="Sales",
    hole=0.5,
    title="Donut Chart - Region Sales Share"
)
fig.show()

# ------------------------------------------------------------
# 11. AREA CHART
# ------------------------------------------------------------
fig = px.area(
    sales_df,
    x="Month",
    y="Sales",
    color="Category",
    title="Area Chart Example"
)
fig.show()

# ------------------------------------------------------------
# 12. HEATMAP
# ------------------------------------------------------------
pivot_data = sales_df.pivot_table(
    index="Region",
    columns="Category",
    values="Sales",
    aggfunc="sum"
)

fig = px.imshow(
    pivot_data,
    text_auto=True,
    title="Heatmap of Sales by Region and Category"
)
fig.show()

# ------------------------------------------------------------
# 13. TREEMAP
# ------------------------------------------------------------
fig = px.treemap(
    sales_df,
    path=["Region", "Category"],
    values="Sales",
    title="Treemap of Sales"
)
fig.show()

# ------------------------------------------------------------
# 14. SUNBURST CHART
# ------------------------------------------------------------
fig = px.sunburst(
    sales_df,
    path=["Region", "Category"],
    values="Sales",
    title="Sunburst Chart of Sales"
)
fig.show()

# ------------------------------------------------------------
# 15. BUBBLE CHART
# ------------------------------------------------------------
fig = px.scatter(
    df_gapminder.query("year == 2007"),
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    size_max=60,
    title="Bubble Chart - Gapminder 2007"
)
fig.show()

# ------------------------------------------------------------
# 16. FACET CHARTS
# ------------------------------------------------------------
fig = px.scatter(
    df_iris,
    x="sepal_width",
    y="sepal_length",
    color="species",
    facet_col="species",
    title="Facet Scatter Plot"
)
fig.show()

# ------------------------------------------------------------
# 17. ANIMATION
# ------------------------------------------------------------
fig = px.scatter(
    df_gapminder,
    x="gdpPercap",
    y="lifeExp",
    animation_frame="year",
    animation_group="country",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    size_max=55,
    title="Animated Gapminder Chart"
)
fig.show()

# ------------------------------------------------------------
# 18. MAP PLOT (SCATTER GEO)
# ------------------------------------------------------------
india_cities = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Bhopal", "Bengaluru", "Kolkata"],
    "Lat": [28.61, 19.07, 23.25, 12.97, 22.57],
    "Lon": [77.20, 72.87, 77.41, 77.59, 88.36],
    "Sales": [500, 700, 300, 650, 450]
})

fig = px.scatter_geo(
    india_cities,
    lat="Lat",
    lon="Lon",
    size="Sales",
    hover_name="City",
    title="Sales by City on Map"
)
fig.show()

# ------------------------------------------------------------
# 19. 3D SCATTER PLOT
# ------------------------------------------------------------
fig = px.scatter_3d(
    df_iris,
    x="sepal_length",
    y="sepal_width",
    z="petal_length",
    color="species",
    title="3D Scatter Plot"
)
fig.show()

# ------------------------------------------------------------
# 20. 3D LINE PLOT
# ------------------------------------------------------------
t = np.linspace(0, 10, 200)
df_3d = pd.DataFrame({
    "x": np.sin(t),
    "y": np.cos(t),
    "z": t
})

fig = px.line_3d(
    df_3d,
    x="x",
    y="y",
    z="z",
    title="3D Line Plot"
)
fig.show()

# ============================================================
# 21. CUSTOMIZATION IN PLOTLY
# ============================================================

# ------------------------------------------------------------
# 21.1 TITLES, LABELS, TEMPLATE
# ------------------------------------------------------------
fig = px.bar(
    sales_df,
    x="Month",
    y="Sales",
    title="Customized Bar Chart",
    template="plotly_dark"
)

fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Sales Amount",
    title_x=0.5,  # center title
    font=dict(size=14)
)

fig.show()

# ------------------------------------------------------------
# 21.2 CHANGING COLORS
# ------------------------------------------------------------
fig = px.bar(
    sales_df,
    x="Month",
    y="Sales",
    color="Category",
    color_discrete_sequence=px.colors.qualitative.Set2,
    title="Bar Chart with Custom Colors"
)
fig.show()

# ------------------------------------------------------------
# 21.3 MARKERS + LINE STYLE
# ------------------------------------------------------------
fig = px.line(
    sales_df,
    x="Month",
    y="Sales",
    markers=True,
    title="Line Chart with Markers"
)

fig.update_traces(
    line=dict(width=4, dash="dash"),
    marker=dict(size=10, symbol="circle")
)

fig.show()

# ------------------------------------------------------------
# 21.4 HOVER CUSTOMIZATION
# ------------------------------------------------------------
fig = px.scatter(
    sales_df,
    x="Sales",
    y="Profit",
    size="Orders",
    color="Region",
    title="Customized Hover"
)

fig.update_traces(
    hovertemplate="<b>Sales:</b> %{x}<br><b>Profit:</b> %{y}<br><extra></extra>"
)

fig.show()

# ------------------------------------------------------------
# 21.5 LEGEND CUSTOMIZATION
# ------------------------------------------------------------
fig = px.bar(
    sales_df,
    x="Region",
    y="Sales",
    color="Category",
    title="Legend Customization"
)

fig.update_layout(
    legend_title="Product Category",
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig.show()

# ------------------------------------------------------------
# 21.6 ANNOTATIONS
# ------------------------------------------------------------
fig = px.line(
    sales_df,
    x="Month",
    y="Sales",
    title="Line Chart with Annotation"
)

max_idx = sales_df["Sales"].idxmax()
max_month = sales_df.loc[max_idx, "Month"]
max_sales = sales_df.loc[max_idx, "Sales"]

fig.add_annotation(
    x=max_month,
    y=max_sales,
    text="Highest Sales",
    showarrow=True,
    arrowhead=2
)

fig.show()

# ------------------------------------------------------------
# 21.7 SHAPES (Horizontal / Vertical Lines)
# ------------------------------------------------------------
fig = px.line(sales_df, x="Month", y="Sales", title="Shapes Example")

fig.add_hline(y=sales_df["Sales"].mean(), line_dash="dash", annotation_text="Average Sales")
fig.add_vline(x=sales_df["Month"].iloc[5], line_dash="dot", annotation_text="Mid-Year")

fig.show()

# ============================================================
# 22. GRAPH_OBJECTS (MORE CONTROL)
# ============================================================
# plotly.graph_objects = full manual control over chart elements

# ------------------------------------------------------------
# 22.1 SIMPLE GO BAR CHART
# ------------------------------------------------------------
fig = go.Figure()

fig.add_trace(go.Bar(
    x=sales_df["Month"],
    y=sales_df["Sales"],
    name="Sales"
))

fig.update_layout(
    title="Bar Chart using graph_objects",
    xaxis_title="Month",
    yaxis_title="Sales"
)

fig.show()

# ------------------------------------------------------------
# 22.2 MULTIPLE TRACES
# ------------------------------------------------------------
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=sales_df["Month"],
    y=sales_df["Sales"],
    mode="lines+markers",
    name="Sales"
))

fig.add_trace(go.Scatter(
    x=sales_df["Month"],
    y=sales_df["Profit"],
    mode="lines+markers",
    name="Profit"
))

fig.update_layout(title="Sales vs Profit using graph_objects")
fig.show()

# ------------------------------------------------------------
# 22.3 SECONDARY Y-AXIS
# ------------------------------------------------------------
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Bar(x=sales_df["Month"], y=sales_df["Sales"], name="Sales"),
    secondary_y=False
)

fig.add_trace(
    go.Scatter(x=sales_df["Month"], y=sales_df["Profit"], name="Profit"),
    secondary_y=True
)

fig.update_layout(title="Sales and Profit with Secondary Axis")
fig.update_yaxes(title_text="Sales", secondary_y=False)
fig.update_yaxes(title_text="Profit", secondary_y=True)

fig.show()

# ============================================================
# 23. SUBPLOTS
# ============================================================
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Sales Trend", "Profit Trend", "Sales by Region", "Profit Histogram")
)

# Plot 1
fig.add_trace(
    go.Scatter(x=sales_df["Month"], y=sales_df["Sales"], mode="lines+markers", name="Sales"),
    row=1, col=1
)

# Plot 2
fig.add_trace(
    go.Scatter(x=sales_df["Month"], y=sales_df["Profit"], mode="lines+markers", name="Profit"),
    row=1, col=2
)

# Plot 3
region_sum = sales_df.groupby("Region", as_index=False)["Sales"].sum()
fig.add_trace(
    go.Bar(x=region_sum["Region"], y=region_sum["Sales"], name="Region Sales"),
    row=2, col=1
)

# Plot 4
fig.add_trace(
    go.Histogram(x=sales_df["Profit"], name="Profit Distribution"),
    row=2, col=2
)

fig.update_layout(height=700, width=1000, title_text="Subplots Example")
fig.show()

# ============================================================
# 24. DROPDOWN INTERACTIVITY
# ============================================================
fig = go.Figure()

fig.add_trace(go.Bar(x=sales_df["Month"], y=sales_df["Sales"], name="Sales"))
fig.add_trace(go.Bar(x=sales_df["Month"], y=sales_df["Profit"], name="Profit"))

fig.update_layout(
    title="Interactive Dropdown Example",
    updatemenus=[
        {
            "buttons": [
                {
                    "label": "Show Sales",
                    "method": "update",
                    "args": [{"visible": [True, False]},
                             {"title": "Sales Only"}]
                },
                {
                    "label": "Show Profit",
                    "method": "update",
                    "args": [{"visible": [False, True]},
                             {"title": "Profit Only"}]
                },
                {
                    "label": "Show Both",
                    "method": "update",
                    "args": [{"visible": [True, True]},
                             {"title": "Sales and Profit"}]
                }
            ],
            "direction": "down",
            "showactive": True
        }
    ]
)

fig.show()

# ============================================================
# 25. SLIDER EXAMPLE
# ============================================================
years = sorted(df_gapminder["year"].unique())
fig = go.Figure()

for year in years:
    df_year = df_gapminder[df_gapminder["year"] == year]
    fig.add_trace(
        go.Scatter(
            x=df_year["gdpPercap"],
            y=df_year["lifeExp"],
            mode="markers",
            marker=dict(size=8),
            name=str(year),
            text=df_year["country"],
            visible=(year == years[0])
        )
    )

steps = []
for i, year in enumerate(years):
    step = dict(
        method="update",
        args=[
            {"visible": [j == i for j in range(len(years))]},
            {"title": f"Gapminder Data for {year}"}
        ],
        label=str(year)
    )
    steps.append(step)

sliders = [dict(active=0, currentvalue={"prefix": "Year: "}, steps=steps)]

fig.update_layout(
    sliders=sliders,
    title=f"Gapminder Data for {years[0]}",
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy"
)

fig.show()

# ============================================================
# 26. DASHBOARD-LIKE KPI + CHART COMBO
# ============================================================
total_sales = sales_df["Sales"].sum()
total_profit = sales_df["Profit"].sum()
total_orders = sales_df["Orders"].sum()

fig = make_subplots(
    rows=2, cols=2,
    specs=[
        [{"type": "indicator"}, {"type": "indicator"}],
        [{"type": "indicator"}, {"type": "xy"}]
    ],
    subplot_titles=("Total Sales", "Total Profit", "Total Orders", "Sales Trend")
)

# KPI 1
fig.add_trace(go.Indicator(
    mode="number",
    value=total_sales,
    title={"text": "Total Sales"}
), row=1, col=1)

# KPI 2
fig.add_trace(go.Indicator(
    mode="number",
    value=total_profit,
    title={"text": "Total Profit"}
), row=1, col=2)

# KPI 3
fig.add_trace(go.Indicator(
    mode="number",
    value=total_orders,
    title={"text": "Total Orders"}
), row=2, col=1)

# Sales trend
fig.add_trace(go.Scatter(
    x=sales_df["Month"],
    y=sales_df["Sales"],
    mode="lines+markers",
    name="Sales"
), row=2, col=2)

fig.update_layout(height=700, title="Mini Dashboard using Plotly")
fig.show()

# ============================================================
# 27. SAVE PLOT AS HTML
# ============================================================
fig = px.line(sales_df, x="Month", y="Sales", title="Save Plot Example")
fig.write_html("plotly_chart.html")
print("Chart saved as plotly_chart.html")

# ============================================================
# 28. IMPORTANT PLOTLY CONCEPTS YOU MUST KNOW
# ============================================================

print("""
PLOTLY CONCEPTS SUMMARY

1. plotly.express (px)
   - Fast, short syntax
   - Best for quick charts and EDA

2. plotly.graph_objects (go)
   - More control
   - Best for dashboards and custom visuals

3. make_subplots()
   - Used to place multiple charts in one figure

4. fig.update_layout()
   - Used to change title, legend, theme, axis labels, margins, etc.

5. fig.update_traces()
   - Used to change bars, lines, markers, hover behavior, etc.

6. fig.add_annotation()
   - Add text labels / notes on chart

7. fig.add_hline(), fig.add_vline()
   - Add reference lines

8. template=
   - plotly
   - plotly_white
   - plotly_dark
   - ggplot2
   - seaborn
   - simple_white

9. Interactivity:
   - hover
   - zoom
   - pan
   - dropdowns
   - sliders
   - animation

10. Export:
   - fig.write_html("file.html")
""")
```
