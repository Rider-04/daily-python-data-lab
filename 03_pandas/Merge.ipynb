```python


                import pandas as pd
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
df1
```

```python
df2=pd.DataFrame({"A":[1,2,3,4,6],"B":[1000,2000,3000,4000,5000]})
df2
```

```python
df=pd.merge(df1,df2,how="cross")
df
```

```python
df=pd.merge(df1,df2,on="A",how="inner")
df
```

```python
df=pd.merge(df1,df2,on="A",how="right")
df
```

```python
df=pd.merge(df1,df2,on="A",how="outer")
df
```

```python
df=pd.merge(df1,df2,on="A",how="inner")
df
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
df1
```

```python
df2=pd.DataFrame({"C":[1,2,3,4,6],"D":[1000,2000,3000,4000,5000]})
df2
```

```python
df=pd.merge(df1,df2,left_on="A",right_on="C",how="inner")
df
```

```python
df=pd.merge(df1,df2,left_on="A",right_on="C",how="left")
df
```

```python
df=pd.merge(df1,df2,left_on="A",right_on="C",how="left")
df
```

```python
df=pd.merge(df1,df2,left_on="A",right_on="C",how="outer")
df
```
