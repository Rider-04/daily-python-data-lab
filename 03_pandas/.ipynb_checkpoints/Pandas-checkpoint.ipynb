```python
import pandas as pd
```

```python
l=[10,20,30,40]
s=pd.Series(l)
print(s)
print(s[0])
```

```python
l=[10,20,30,40]
s=pd.Series(l,index=['a','s','e','d'],dtype='float',name='python')
print(s)
print(s['a'])
```

```python
d={'A':[10,20,30],"B":[100,200,300]}
s=pd.Series(d)
print(s)
```

```python
d={'A':[10,20,30],"B":[100,200,300]}
s=pd.DataFrame(d)
print(s)
for i in range(3):
    if "Sum" not in s:
        s["Sum"]=[]
    s["Sum"].append(s["A"][i]+s["B"][i])
print(s)
```
