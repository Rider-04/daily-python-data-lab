# Domain :- Public Safety Analytics
Traffic accident database consists of all accidents that happened in Slovenia’s capital city Ljubljana between the years 1995 and 2005.

```python
import mysql.connector
import pandas as pd

# =========================
# MYSQL CONNECTION
# =========================

conn = mysql.connector.connect(
    host = "localhost",
    user = "root"      ,
    password = "parth004",
    database = "Accidents"
)

# =========================
# LOAD TABLES INTO DATAFRAMES
# =========================

# Accident Data
nesreca_df = pd.read_sql("SELECT * FROM nesreca", conn)

# Person Data
oseba_df = pd.read_sql("SELECT * FROM oseba", conn)

# Administrative Unit Data
upravna_enota_df = pd.read_sql("SELECT * FROM upravna_enota", conn)

# =========================
# CHECK DATA
# =========================

print("Nesreca Table")
print(nesreca_df.head())

print("\nOseba Table")
print(oseba_df.head())

print("\nUpravna Enota Table")
print(upravna_enota_df.head())

# =========================
# CLOSE CONNECTION
# =========================

conn.close()
```

```python
# =========================
# RENAME TABLE DATAFRAMES
# =========================

accidents_df = nesreca_df.copy()
persons_df = oseba_df.copy()
administrative_units_df = upravna_enota_df.copy()

# =========================
# RENAME COLUMNS
# =========================

# ---------------------------------
# ACCIDENTS TABLE
# ---------------------------------

accidents_df.rename(columns={

    "id_nesreca": "accident_id",
    "klas_nesreca": "accident_class",
    "upravna_enota": "administrative_unit",
    "cas_nesreca": "accident_time",
    "naselje_ali_izven": "settlement_or_outside",
    "kategorija_cesta": "road_category",
    "oznaka_cesta_ali_naselje": "road_or_settlement_code",
    "tekst_cesta_ali_naselje": "road_or_settlement_name",
    "oznaka_odsek_ali_ulica": "section_or_street_code",
    "tekst_odsek_ali_ulica": "section_or_street_name",
    "stacionazna_ali_hisna_st": "station_or_house_number",
    "opis_prizorisce": "scene_description",
    "vzrok_nesreca": "accident_cause",
    "tip_nesreca": "accident_type",
    "vreme_nesreca": "weather_condition",
    "stanje_promet": "traffic_condition",
    "stanje_vozisce": "road_condition",
    "stanje_povrsina_vozisce": "road_surface_condition",
    "x": "x_coordinate",
    "y": "y_coordinate",
    "x_wgs84": "longitude",
    "y_wgs84": "latitude"

}, inplace=True)

# ---------------------------------
# PERSONS TABLE
# ---------------------------------

persons_df.rename(columns={

    "id_nesreca": "accident_id",
    "povzrocitelj_ali_udelezenec": "causer_or_participant",
    "starost": "age",
    "spol": "gender",
    "upravna_enota": "administrative_unit",
    "drzavljanstvo": "citizenship",
    "poskodba": "injury_type",
    "vrsta_udelezenca": "participant_type",
    "varnostni_pas_ali_celada": "seatbelt_or_helmet",
    "vozniski_staz_LL": "driving_experience_years",
    "vozniski_staz_MM": "driving_experience_months",
    "alkotest": "alcohol_test",
    "strokovni_pregled": "medical_exam",
    "starost_d": "age_group",
    "vozniski_staz_d": "driving_experience_group",
    "alkotest_d": "alcohol_test_group",
    "strokovni_pregled_d": "medical_exam_group"

}, inplace=True)

# ---------------------------------
# ADMINISTRATIVE UNITS TABLE
# ---------------------------------

administrative_units_df.rename(columns={

    "id_upravna_enota": "administrative_unit_id",
    "ime_upravna_enota": "administrative_unit_name",
    "st_prebivalcev": "population",
    "povrsina": "area"

}, inplace=True)

# =========================
# CHECK RENAMED COLUMNS
# =========================

print(accidents_df.columns)
print(persons_df.columns)
print(administrative_units_df.columns)
```

```python
acc_cat=list(set(accidents_df.columns)-set(accidents_df.describe().columns))
```

```python
acc_cat
```

```python
accidents_df[acc_cat].sample(10)
```

```python
# ============================================
# FIND UNIQUE VALUES IN ALL CATEGORICAL COLUMNS
# ============================================

# Select categorical/object columns
categorical_columns = accidents_df.select_dtypes(include=["object"]).columns

# Loop through each categorical column
for column in acc_cat:
    
    print("\n" + "="*60)
    print(f"COLUMN NAME : {column}")
    print("="*60)
    
    # Print unique values
    print(accidents_df[column].unique())
    
    # Print total unique count
    print(f"\nTotal Unique Values : {accidents_df[column].nunique()}")
```

```python
# ============================================
# TRANSLATE / MAP CODE VALUES TO ENGLISH
# ============================================

# --------------------------------------------
# SETTLEMENT OR OUTSIDE
# --------------------------------------------

accidents_df["settlement_or_outside"] = accidents_df["settlement_or_outside"].replace({
    
    "D": "Settlement",
    "N": "Outside Settlement"

})

# --------------------------------------------
# SCENE DESCRIPTION
# --------------------------------------------

accidents_df["scene_description"] = accidents_df["scene_description"].replace({

    "C": "Curve",
    "P": "Straight Road",
    "N": "Intersection"

})

# --------------------------------------------
# ACCIDENT CAUSE
# --------------------------------------------

accidents_df["accident_cause"] = accidents_df["accident_cause"].replace({

    "P": "Speeding",
    "N": "Incorrect Driving Side",
    "C": "Improper Movement",
    "O": "Other Causes"

})

# --------------------------------------------
# ACCIDENT TYPE
# --------------------------------------------

accidents_df["accident_type"] = accidents_df["accident_type"].replace({

    "SV": "Vehicle Collision",
    "PV": "Pedestrian Accident",
    "VR": "Rollover Accident",
    "OS": "Side Collision",
    "HI": "Hit and Run"

})

# --------------------------------------------
# ROAD CATEGORY
# --------------------------------------------

accidents_df["road_category"] = accidents_df["road_category"].replace({

    "TV": "Main Road",
    "BT": "Local Road",
    "NT": "Regional Road",
    "OP": "Highway"

})

# --------------------------------------------
# TRAFFIC CONDITION
# --------------------------------------------

accidents_df["traffic_condition"] = accidents_df["traffic_condition"].replace({

    "N": "Normal Traffic",
    "A": "Heavy Traffic",
    "V": "Dense Traffic",
    "M": "Moderate Traffic",
    "5": "Unknown"

})

# --------------------------------------------
# WEATHER CONDITION
# --------------------------------------------

accidents_df["weather_condition"] = accidents_df["weather_condition"].replace({

    "J": "Clear Weather",
    "O": "Cloudy Weather"

})

# --------------------------------------------
# ROAD CONDITION
# --------------------------------------------

accidents_df["road_condition"] = accidents_df["road_condition"].replace({

    "SU": "Dry Road",
    "MO": "Wet Road",
    "SP": "Snowy Road",
    "SL": "Slippery Road"

})

# --------------------------------------------
# ROAD SURFACE CONDITION
# --------------------------------------------

accidents_df["road_surface_condition"] = accidents_df["road_surface_condition"].replace({

    "A": "Asphalt",
    "M": "Macadam",
    "O": "Other Surface"

})

# --------------------------------------------
# ACCIDENT CLASS
# --------------------------------------------

accidents_df["accident_class"] = accidents_df["accident_class"].replace({

    "B": "Minor Accident",
    "U": "Serious Accident"

})

# ============================================
# CHECK OUTPUT
# ============================================

print(accidents_df.head())
```

```python
# =====================================================
# COMPLETE TRANSLATION / LABEL MAPPING
# =====================================================

# -----------------------------------------------------
# SCENE DESCRIPTION
# -----------------------------------------------------

accidents_df["scene_description"] = accidents_df["scene_description"].replace({

    "Curve": "Curve",
    "Straight Road": "Straight Road",
    "Intersection": "Intersection",

    "K": "Junction",
    "R": "Roundabout",
    "Ž": "Railway Crossing",
    "Z": "Crosswalk",
    "A": "Highway Area",
    "M": "Bridge / Tunnel"

})

# -----------------------------------------------------
# ACCIDENT CAUSE
# -----------------------------------------------------

accidents_df["accident_cause"] = accidents_df["accident_cause"].replace({

    "PR": "Speeding",
    "HI": "Hit and Run",
    "SV": "Wrong Driving Direction",
    "PD": "Unsafe Distance",
    "VR": "Improper Overtaking",
    "PV": "Pedestrian Violation",
    "NP": "Road Condition Negligence",
    "OS": "Other Causes",
    "CE": "Vehicle Defect",
    "TO": "Technical Failure",
    "VO": "Driver Fatigue"

})

# -----------------------------------------------------
# ACCIDENT TYPE
# -----------------------------------------------------

accidents_df["accident_type"] = accidents_df["accident_type"].replace({

    "BT": "Vehicle Collision",
    "ÈT": "Chain Collision",
    "TO": "Vehicle Overturn",
    "Side Collision": "Side Collision",
    "NT": "Truck Accident",
    "OP": "Obstacle Collision",
    "TV": "Vehicle-Pedestrian Accident",
    "PR": "Road Departure",
    "PP": "Rear-End Collision",
    "PZ": "Parking Accident"

})

# -----------------------------------------------------
# ROAD CATEGORY
# -----------------------------------------------------

accidents_df["road_category"] = accidents_df["road_category"].replace({

    "V": "Urban Road",
    "L": "Local Road",
    "M": "Main Road",
    "N": "National Road",
    "R": "Regional Road",
    "A": "Highway",
    "1": "Category 1 Road",
    "2": "Category 2 Road",
    "3": "Category 3 Road",
    "4": "Category 4 Road",
    "5": "Category 5 Road",
    "T": "Tourist Road",
    "0": "Unknown Road Type",
    "H": "Expressway"

})

# -----------------------------------------------------
# TRAFFIC CONDITION
# -----------------------------------------------------

accidents_df["traffic_condition"] = accidents_df["traffic_condition"].replace({

    "R": "Regular Traffic",
    "Normal Traffic": "Normal Traffic",
    "G": "Heavy Traffic",
    "E": "Light Traffic",
    "Z": "Traffic Congestion"

})

# -----------------------------------------------------
# WEATHER CONDITION
# -----------------------------------------------------

accidents_df["weather_condition"] = accidents_df["weather_condition"].replace({

    "Cloudy Weather": "Cloudy Weather",
    "Clear Weather": "Clear Weather",

    "M": "Foggy Weather",
    "D": "Rainy Weather",
    "S": "Snowy Weather",
    "N": "Night / Dark",
    "V": "Windy Weather",
    "T": "Stormy Weather"

})

# -----------------------------------------------------
# ROAD CONDITION
# -----------------------------------------------------

accidents_df["road_condition"] = accidents_df["road_condition"].replace({

    "Wet Road": "Wet Road",
    "Dry Road": "Dry Road",
    "Snowy Road": "Snowy Road",
    "Slippery Road": "Slippery Road",

    "PN": "Uneven Road",
    "SN": "Snow Covered",
    "PP": "Damaged Surface",
    "OS": "Other Condition",
    "BL": "Mud / Dirt Road"

})

# -----------------------------------------------------
# ACCIDENT CLASS
# -----------------------------------------------------

accidents_df["accident_class"] = accidents_df["accident_class"].replace({

    "H": "Fatal Accident",
    "Minor Accident": "Minor Accident",
    "L": "Accident with Injuries",
    "S": "Severe Accident",
    "P": "Property Damage Only",
    "Serious Accident": "Serious Accident"

})

# =====================================================
# CHECK CLEANED DATA
# =====================================================

print(accidents_df.head())
```
