```python
def fillnulls(dataframe):
    column=input("Enter Column Name:-")
    Q1= np.quantile(dataframe[column].dropna(),0.25)
    Q3= np.quantile(dataframe[column].dropna(),0.75)
    IQR=Q3-Q1
    lower_fence=Q1-(1.5*IQR)
    upper_fence=Q3+(1.5*IQR)
```
