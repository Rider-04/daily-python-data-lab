# Operators

### Arithmetic Operators

```python
# +,-,*,/,//,**

a=10
b=2
print("Addition:- ",a+b)
print("Substraction:- ",a-b)
print("Multiplication:- ",a*b)
print("Simple Division:- ",a/b)
print("Floor Division:- ",a//b)
print("Modulus:- ",a%b)
print("Exponential Power:- ",a**b)
```

### Assignment Operators

```python
# =, ShortHands(+=,-=,*=,/=,//=,%=,**=)
a=10
b=5
c=a+b
a*=5 # a=a*5
b+=c # b=b+c
c-=a # c=c-a
c=a+b
print(a,'',b,'',c)
```

### Comparison Operator

```python
# ==,!=,>=,<=,>,<
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a<b)
print(a>b)
```

### Logical Operators

```python
# and ,or, not
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>b and not b>c)
```

### Identity Operators

```python
# is,is not
# id()- returns memoru address of an object
a=10
b=8
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```

### Membership Operators

```python
# in , not in
a="hello"
print("H" in a)
print("z" in a)
print("1" in a)
print("e" in a)
print("H" not in a)
```

### Bitwise Operators

```python
# &, |, ^
# bin() - return binary of an integer object
a=42
b=23
print(bin(a))
print('',bin(b))
print(bin(a&b))
print()
print(a&b)
print(a|b)
print(a^b)
```
